```python
import pickle

model = pickle.load(open("/content/drive/Shareddrives/Grupo 4 SSC-5972/Checkpoint 1/CÃ³digos Python/module_health_classifier.pkl", "rb"))
```

```python
import paho.mqtt.client as mqtt

import json

import numpy as np

# The callback for when the client receives a CONNACK response from the server.
def on_connect(client, userdata, flags, rc):
    print("Connected with result code "+str(rc))

    # Subscribing in on_connect() means that if we lose the connection and
    # reconnect then subscriptions will be renewed.
    client.subscribe("16153060/status") #good

# The callback for when a PUBLISH message is received from the server.
def on_message(client, userdata, msg):

    json_data = json.loads(msg.payload)
    print(json_data)

    element = [json_data['chipid'],json_data['minfreeheap'],json_data['freeheap'],json_data['internal_temperature'],json_data['wifi_level'],json_data['reset_reason1'],json_data['reset_reason2']]
    print(model.predict(np.array(element).reshape(1, -1)))
    print("")


client = mqtt.Client()
client.on_connect = on_connect
client.on_message = on_message

client.connect("broker.mqtt-dashboard.com", 1883, 60)

# Blocking call that processes network traffic, dispatches callbacks and
# handles reconnecting.
# Other loop*() functions are available that give a threaded interface and a
# manual interface.
client.loop_forever()
```
